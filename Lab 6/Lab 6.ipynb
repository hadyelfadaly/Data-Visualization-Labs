```python
# Understanding the different data types can help us identify correct preprocessing techniques & convert the data appropriately.
# Furthermore, it will also enable us to perform the best visualizations and uncover hidden knowledge.

#Types of data based on structure
# 1- Structured data: This type of data is usually composed of numbers or words. They are usually stored in Relational databases
# 2- Unstructured data: Including texts, images, videos, speech/audio, and so on.
# 3- Semi-structured: Use of metadata or tags that provide additional information about the data elements. For example (XML)

#Types of data
# 1- Numeric/Quantitative data -> Discrete or continous
# Discrete: Data takes on discrete values or whole numbers i.e. numbers without decimal points
# Continuous: Data takes on integer values i.e. numbers with decimal values.
# 2- Categorical/Qualitative data: This encompasses data that can be represented through words. It usually defines groups or categories. -> oridnal or nominal
# Ordinal: This type of data has an inherent ordering present within the categories.
# Nominal: This type of data has categories that don’t have any particular order or ranking associated with them. The total number of categories is usually finite in this type of data as well.

# Data Formats
# 1- Tabular data: The tabular data consists of multiple features/columns with each of them having a particular data type.
# 2- Time series data: Time series data consists of data points that are indexed at specific points in time.
```

```python
# What is data preprocessing?
# The fundamental purpose of data preparation is to manipulate and transform raw data so that the information content enfolded in the data set
# can be exposed or made more easily accessible.

# Why Data Preprocessing?
# 1- Real data could be dirty and could drive to the extraction of useless patterns/rules. (Incomplete data, data with noise, inconsistent data)
# 2- Data preprocessing can generate a smaller dataset than the original, which allows us to improve the efficiency in the Data Mining process.
# data reduction techniques like:Feature selection, sampling or instance selection and discretization
# 3- No quality data, no quality mining results!

# Major Tasks in Data Preprocessing
# 1- Data integration. Fusion of multiple sources in a Data Warehousing
# 2- Data cleaning. Removal of noise and inconsistencies
# 3- Missing values imputation
# 4- Data Transformation
# 5- Data reduction

# Data Integration
# 1- Obtain data from different information sources
# 2- Address problems of codification and representation.
# 3- Integrate data from different tables to produce homogeneous information

# Data cleaning
# 1- Fix inconsistencies
# 2- Fill/impute missing values
# 3- Smooth noisy data
# 4- Identify or remove outliers

# Data Transformation: To transform data in the best way possible to the application of Data Mining algorithms.
# 1- Aggregation.
# 2- Data generalization. It is to obtain higher degrees of data from the currently available, by using concept hierarchies. (streets -> cities)
# 3- Normalization and Standardization (To convert the values of an attribute to a better range.)
# 4- Lineal transformations, quadratic, polynomial, …
```
