```python
import folium #gives you a map
import seaborn as sns
from branca.element import Figure
import matplotlib as plt
```

```python
data = sns.load_dataset('tips')

data
```

```python
# The regression plots in seaborn are primarily intended to add a visual guide that helps to emphasize patterns in a dataset during exploratory data analyses.
# Regression plots as the name suggests creates a regression line between 2 parameters and helps to visualize their linear relationships.

sns.set_style('whitegrid')
sns.lmplot(x = 'total_bill', y = 'tip', data = data) #the shading is the confidence interval
```

```python
# regression plot with more parameters

sns.lmplot(x ='total_bill', y='tip', data = data, col = 'sex', row = 'time', hue = 'smoker') # hue adds color variable based on the smoker
```

```python
# Folium is a Python library used for visualizing geospatial data

folium.Map((26.8206, 30.8025)) #egypt's map
```

```python
folium.Map((26.8206, 30.8025), width = 550, length = 350, min_zoom =8, max_zoom=14) #the width and length for a sketch we can use beside the map
```

```python
# layers and tiles
# Each tileset shows different features of a map and is suitable for different purposes.

fig = Figure(width = 550, height = 350)
m = folium.Map(location=[26.8206, 30.8025])

fig.add_child(m)
folium.TileLayer('cartodbpositron').add_to(m)
folium.TileLayer('cartodbdark_matter').add_to(m)
folium.LayerControl().add_to(m)
m
```

```python
# Plotting Markers on the map

#Adding markers to the map
folium.Marker(location=[28.695800, 77.244721],popup='Default popup Marker1',tooltip='Clickhere to see Popup').add_to(m)
folium.Marker(location=[28.645800,77.214721],popup='<strong>Marker3</strong>',tooltip='<strong>Click here to see Popup</strong>').add_to(m)
folium.Marker(location=[28.655800, 77.274721],popup='<h3style="color:green;">Marker2</h3>',tooltip='<strong>Click here to see Popup</strong>').add_to(m)

m

# <strong> to appear bold
```

```python
# icons: Icon() takes three arguments â€“ color, prefix and icon

```

```python
# exercise 

cairoMap = folium.Map([30.0444, 31.2357], zoom_start = 12)

folium.Marker(location=[30.0444, 31.2357], popup='<strong>Tahrir Square</strong>', icon=folium.Icon(color='red', icon = 'none')).add_to(cairoMap)
folium.Marker(location=[29.9792, 31.1342], popup='<strong>Pyramids of Giza</strong>', icon=folium.Icon(color='green', icon = 'none')).add_to(cairoMap)
folium.Marker(location=[30.0295, 31.2615], popup='<strong>Citadel of Cairo</strong>').add_to(cairoMap)

museums = folium.FeatureGroup(name='Museums')

folium.Marker(location=[30.0478, 31.2336], popup='<strong>Egyptian Museum</strong>', icon=folium.Icon(color='gray', icon = 'none')).add_to(museums)
folium.Marker(location=[30.0474, 31.2294], popup='<strong>Museum of Islamic Art</strong>', icon=folium.Icon(color='black', icon = 'none')).add_to(museums)
museums.add_to(cairoMap)

parks = folium.FeatureGroup(name='Parks')

folium.Marker(location=[30.0326, 31.2645], popup='<strong>Al-Azhar Park</strong>', icon=folium.Icon(color='lightblue', icon = 'none')).add_to(parks)
folium.Marker(location=[29.9682, 31.2695], popup='<strong>International Garden</strong>', icon=folium.Icon(color='white', icon = 'none')).add_to(parks)
parks.add_to(cairoMap)

folium.TileLayer('cartodbpositron').add_to(cairoMap)
folium.TileLayer('cartodbdark_matter').add_to(cairoMap)
folium.LayerControl().add_to(cairoMap)

cairoMap
```
