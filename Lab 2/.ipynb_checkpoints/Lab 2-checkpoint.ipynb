```python
# NumPy offers the random module to work with random numbers
from numpy import random
```

```python
x = random.randint(100) #random int from 0 to 100
y = random.rand() #random float from 0 to 1
z = random.randint(100, size=(3, 5)) #Generate a 2-D array with 3 rows, each row containing 5 random integers from 0 to 100:

print(x)
print(y)
print(z)
```

```python
# random data distribution

# Data Distribution is a list of all possible values, and how often each value occurs.
# The random module offer methods that returns randomly generated data distributions.

# Random Data Distribution: is a set of random numbers that follow a certain probability density function.
# Probability Density Function: A function that describes a continuous probability. i.e. probability of all values in an array.

# We can generate random numbers based on defined probabilities using the choice() method of the random module.

# Generate a 1-D array containing 100 values, where each value has to be 3, 5, 7 or 9.
# The probability for the value to be 3 is set to be 0.1
# The probability for the value to be 5 is set to be 0.3
# The probability for the value to be 7 is set to be 0.6
# The probability for the value to be 9 is set to be 0

x = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(100))

print(x)

# Exercise

y = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(3, 5))

print(y)

# random normal data distribution

z = random.normal(loc = 4, scale = 10, size = (3, 5)) #loc = mean, scale - standard deviation

print(z)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Distplot stands for distribution plot, it takes as input an array and plots a curve corresponding to the distribution of points in the array.

sns.distplot(z)

plt.show()
```

```python
# data cleaning

data = pd.read_csv('data_new.csv')

data.info() # there exist some null values in some rows
```

```python
# data cleaning exercise

data['Calories'].fillna(data['Calories'].mean(), inplace = True)
```

```python
data.info() #nulls are filled
```

```python
# fixing wrong format issue

data['Date'] = pd.to_datetime(data['Date'], format='mixed')

print(data.to_string())
```

```python
# wrong data exercise

for x in data.index:

    if data.loc[x, 'Duration'] > 120:
        data.drop(x, inplace=True)

print(data)
```

```python
# remove all duplicates

data.drop_duplicates(inplace=True) #data.duplicated() return true for every row that is a duplicate

print(data)
```

```python
# correlations, Finding relationships between each column in the data set

print(data.corr()) # The corr() method ignores "not numeric" columns

# results vary from -1 to 1
# 1 means perfect correlation, means each time a value went up the other went up to
# -1 absolutley opposite to eachother
# the nearer you are to 0 means no correlation
```
