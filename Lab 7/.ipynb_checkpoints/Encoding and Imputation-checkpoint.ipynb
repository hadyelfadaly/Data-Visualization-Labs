```python
import pandas as pd
import seaborn as sns
import category_encoders as ce
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
loan_data = pd.read_csv(r'D:\FCI\Semster 1\Data Visualization\Fall 2025\Lectures\Lab6\Loan_Default.csv')
loan_data.drop(['ID', 'year'], axis=1, inplace =True)
```

```python
categorical_features = loan_data.select_dtypes(include=['object']).columns.tolist()
```

```python
Ordinal_features = ['age']
Nominal_features = categorical_features.copy()
Nominal_features.remove('age')
```

### Ordinal Data Encoding

```python
enc = OrdinalEncoder()
loan_data[Ordinal_features] = enc.fit_transform(loan_data[Ordinal_features])
```

### Nominal Data Encoding

```python
# category_encoders Library 
loan_data1 = loan_data.copy()
OHE = ce.OneHotEncoder(cols=Nominal_features, handle_unknown='return_nan', return_df=True, use_cat_names=True)

encoded_categories = OHE.fit_transform(loan_data1[Nominal_features])

loan_data1.drop(Nominal_features, axis=1, inplace=True)
loan_data1 = pd.concat([loan_data1, encoded_categories], axis=1)
```

```python
#SKLearn Library
loan_data2 = loan_data.copy()
OHE = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

encoded_categories = pd.DataFrame(OHE.fit_transform(loan_data2[Nominal_features]))

loan_data2.drop(Nominal_features, axis=1, inplace=True)
loan_data2 = pd.concat([loan_data2, encoded_categories], axis=1)
```

```python
# Dummy Encoding
loan_data3 = loan_data.copy()

encoded_categories = pd.get_dummies(data=loan_data3[Nominal_features], drop_first=True)

loan_data3.drop(Nominal_features, axis=1, inplace=True)
loan_data3 = pd.concat([loan_data3, encoded_categories], axis=1)
```

```python
# frequency Encoding
loan_data4 = loan_data.copy()

for c in Nominal_features:
    frequency = loan_data4[c].value_counts(normalize=True)
    loan_data4[c+'_freq'] = loan_data4[c].map(frequency)

loan_data4 =  loan_data4.drop(Nominal_features, axis=1)
```

```python
# Target Encoding
loan_data5 = loan_data.copy()

TE = ce.TargetEncoder(cols=Nominal_features)

loan_data5 = TE.fit_transform(loan_data5, loan_data5['loan_amount'])
```

### Missing Values Imputation

```python
loan_data4.isna().sum()
```

```python
high_missing_cols = ['rate_of_interest', 'Interest_rate_spread', 'Upfront_charges', 'property_value', 'LTV', 'dtir1']
cols_to_be_imputed = ['term', 'income', 'age', 'loan_limit_freq', 'approv_in_adv_freq', 'loan_purpose_freq', 'Neg_ammortization_freq', 'submission_of_application_freq']
```

```python
# Drop Columns with high missing values
loan_data4.drop(high_missing_cols,axis=1,inplace=True)
```

```python
loan_data4[loan_data4[cols_to_be_imputed].isnull().any(axis=1)][cols_to_be_imputed]
```

```python
# Constant Imputer 
loan_data6 = loan_data4.copy()
const_imputer = SimpleImputer(strategy='constant', fill_value=99999)
loan_data6[cols_to_be_imputed] = const_imputer.fit_transform(loan_data6[cols_to_be_imputed])
```

```python
# Mode Imputer
loan_data7 = loan_data4.copy()
const_imputer = SimpleImputer(strategy='most_frequent')
loan_data7[cols_to_be_imputed] = const_imputer.fit_transform(loan_data7[cols_to_be_imputed])
```

```python
# mean/median
loan_data8 = loan_data4.copy()
for c in cols_to_be_imputed:
    loan_data8[c].fillna(loan_data8[c].mean(), inplace=True)
```
