```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# basic plotting exercise

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints) #When you plot only one array (plt.plot(y)), Matplotlib uses the array’s indices [0, 1, 2, ...] as the default X-values.
plt.show()
```

```python
# modify linestyle to dotted line

plt.plot(ypoints, linestyle = 'dotted') # or linestyle = ‘dashed’, plt.plot(ypoints, ls = ':') is a shorter syntax
plt.show()

# dashed = --
# dotted = :
# solid (default) = -
# dashdot = -.
# none = '' (leaves the graph empty)
```

```python
# modify line color

plt.plot(ypoints, color = 'g') #or c = ‘g’
plt.show()
```

```python
# modify linewidth

plt.plot(ypoints, linewidth = '20.5') #or lw = ’20.5’
plt.show()
```

```python
# multiple lines exercise

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

#plt.plot(x1, y1, x2, y2) #if we wanna modify width, color or style we have to seperate each line alone like this

# arguments order doesnt matter
plt.plot(x1, y1, ls = ':', c = 'g', lw = '10')
plt.plot(x2, y2, c = 'b', ls = '--')
plt.show()
```

```python
# bar plots

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y)
plt.show()
```

```python
# bar plots exercise

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, color = 'r', height= 0.2)
plt.show()
```

```python
# scatter plot

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

```python
# scatter plot exercise2

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size = (100))
sizes = 10 * np.random.randint(100, size = (100))

plt.scatter(x, y, c = colors, s = sizes, alpha = 0.5, cmap = 'nipy_spectral') #cmap maps numbers into colors, alpha is transperancy ratio
plt.colorbar() # shows the color bar on the left side
plt.show()
```

```python
# bubble plot: is a scatterplot where the circle size is mapped to the value of a third numeric variable.

price = [600, 250, 45, 120, 80, 450]
rating = [4.7, 4.5, 4.8, 4.3, 3.8, 4.6]
units_sold = [800, 4500, 15000, 2500, 3000, 1200]

# Bubble sizes scaled (so they’re visible)
bubble_sizes = [p * 0.1 for p in units_sold]

plt.scatter(price, rating, s=bubble_sizes, alpha=0.4, c='r')
plt.show()
```

```python
# box plots: is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum.

np.random.seed(42)

dataSet = np.random.normal(100, 25, 200)

figure = plt.figure(figsize=(10, 8))
    
# plt.boxplot(dataSet) or
plt.boxplot(dataSet,vert=False,patch_artist=True,notch='False') #notch=True adds a confidence interval around the median in a box plot — useful for quickly comparing medians across groups.
plt.show()
```

```python
# boxplot exercise

np.random.seed(10)

dataSet1 = np.random.normal(100, 10, 220)
dataSet2 = np.random.normal(80, 20, 200)
dataSet3 = np.random.normal(60, 35, 220)
dataSet4 = np.random.normal(50, 40, 200)

dataSet = [dataSet1, dataSet2, dataSet3, dataSet4]
plt.boxplot(dataSet)
    
plt.show()
```

```python
# waffle charts: It is especially useful to display the progress towards goals and seeing how each item contributes to the whole.
from pywaffle import Waffle
import pandas as pd

df = pd.DataFrame({
'country': ['Argentina', 'Brazil', 'Cuba', 'Peru'],
'number': [212, 334, 1500, 130]
})

fig = plt.figure(
FigureClass=Waffle,
rows=7,
values=list(df.number/5),
labels=list(df.country),
figsize=(12, 8))
```

```python
# wordclouds: It is basically a visualization technique to represent the frequency of words in a text where the size of the word represents its frequency.

from wordcloud import WordCloud, STOPWORDS

df = pd.read_csv('Restaurant_Reviews.tsv', sep='\t')
text = "".join(word for word in df.Review)
word_cloud = WordCloud(width=3000, height=2000, random_state=1, background_color='white', colormap='viridis',
collocations=False, stopwords=STOPWORDS).generate(text)

# reandom_state: To recreate the same plot every time we run the function. The random_state parameter has to be an integer value.
# collocations: To include bigrams of two words when set to True. The default value is True
# stopwords: To set the list of words that needs to be eliminated. This list can include trivial words like this, that, is, was, the,etc.
# If this parameter is set to None, then function will consider a built-in list of STOPWORDS.

plt.imshow(word_cloud)
plt.axis("off")
plt.show()
```
